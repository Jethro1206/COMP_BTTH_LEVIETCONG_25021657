```python
#W8A3
import sys

try:
  n = int(input())
except ValueError:
  print("invalid input")


exists = {}

def happy(n):

  x, sum = n, 0
  while x > 0:
    sum += int(x % 10) * int(x % 10)
    x //= 10

  # print(sum)
  if sum == 1:
    print("YES")
    # sys.exit()
  elif sum in exists or sum == n:
    print("NO")
    # sys.exit()
  else:
    exists[sum] = 1
    happy(sum)

happy(n)
```

```python
#W8A10

try:
    a = list(map(int, input().split()))
    n = a[0]
    m = a[1:]

    if len(m) != n or any(x <= 0 for x in m) or len(set(m)) != len(m):
        raise ValueError

    print(sum(m))
except:
    print("Mang khong hop le")
```

```python
#W8A11

import re

try:
    t = int(input())
    for _ in range(t):
        s = input()
        try:
            re.compile(s)
            print("True")
        except re.error:
            print("False")
except ValueError:
    pass
```
