```python
#3
import math
a = int(input())
def isPrime(a):
  if a < 2:
    return False
  for i in range(2, int(math.sqrt(a)) + 1):
    if a % i == 0:
      return False
  return True

print(isPrime(a))
```

```python
#4
a = int(input())

sum = 0
for i in range(1, a):
  if a % i == 0:
    sum += i

if sum == a:
  print(True)
else:
  print(False)
```

```python
#6
a = int(input())
def factorial(a):
  fac = 1
  for i in range(1, a + 1):
    fac *= i
  print(fac)
factorial(a)
```

```python
#7
def calculate(a,b,k):
    if k=='+':
        return a+b
    elif k=='-':
        return a-b
    elif k=='*':
        return a*b
    elif k=='/':
        return a/b
a,k,b=map(str,input().split())
a=float(a)
b=float(b)
print(f"{calculate(a, b, k):.2f}")
```

```python
#8
a, b = map(int, input().split())

def xuly(a, b):
  a1 = ""
  while a > 0:
    a1 = a1 + str(int(a % 2))
    a //= 2
  a2 = ""
  a1 = a1[::-1]
  while b > 0:
    a2 = a2 + str(int(b % 2))
    b //= 2

  a2 = a2[::-1]

  if len(a1) > len(a2):
    a1, a2 = a2, a1

  n = len(a1)
  m = len(a2)
  while(n < m):
    a1 = "0" + a1
    n += 1

  cnt = 0
  for i in range(min(len(a1), len(a2))):
    if a1[i] != a2[i]:
      cnt += 1
  print(cnt)
xuly(a, b)
```

```python
#10
def check(s1,s2):
    if len(s1) != len(s2):
        return False
    a=[s1[1]]
    b=[s2[1]]
    k=[0]
    cnt=1
    for i in range(1,len(s1)):
        if s1[i] in a:
            for j in range(cnt):
                if a[j] == s1[i]:
                    if b[j] != s2[i]:
                        return False
        elif s2[i] in b:
            for j in range(cnt):
                if b[j] == s2[i]:
                    if b[j] != s1[i]:
                        return False
        elif s1[i] not in a and s2[i] not in b:
            a.append(s1[i])
            b.append(s2[i])
            cnt+=1
    return True
a,b=map(str,input().split())
print(check(a,b))
```
