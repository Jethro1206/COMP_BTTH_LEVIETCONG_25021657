```python
#W6A3
def rotate(t, k):
    if not t:
        return ()

    n = len(t)
    k = k % n

    rotated_list = list(t[k:]) + list(t[:k])

    return tuple(rotated_list)

original_tuple = tuple(map(int, input().split()))
k = int(input())

result_tuple = rotate(original_tuple, k)

print(result_tuple)
```

```python
#W6A4
from collections import defaultdict

def c(s):
    d = defaultdict(list)
    p = s.split()

    for item in p:
      k, v = item.split(':', 1)
      d[k].append(v)


    return dict(d)

s = input()
r = c(s)
print(r)
```

```python
#W6A8
def c(l):
    d = {}
    for w in l:
        if w in d:
            d[w] += 1
        else:
            d[w] = 1
    return d

s = input()
l = s.split()
r = c(l)
print(r)
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())

r = {}

k = set(d1.keys()) | set(d2.keys())

for key in k:
    v1 = d1.get(key, 0)
    v2 = d2.get(key, 0)
    r[key] = v1 + v2

s = dict(sorted(r.items()))
print(s)
```

```python
#W6A10
def f(l, k):
    r = []
    n = len(l)

    for i in range(n):
        for j in range(i, n):
            if l[i] + l[j] == k:
                p = (l[i], l[j])
                r.append(p)

    r.sort()
    return r

s = input()
l = list(map(int, s.split()))
k = int(input())

rs = f(l, k)
print(rs)
```

```python
#W6A11
def t(t):
    e = []
    o = []

    for x in t:
        if x % 2 == 0:
            e.append(x)
        else:
            o.append(x)

    e_t = tuple(e)
    o_t = tuple(o)

    return e_t, o_t

s = input()
t_in = tuple(map(int, s.split()))

e_out, o_out = t(t_in)
print(e_out)
print(o_out)
```

```python
#W6A12
def f(l):
    c = {}
    for x in l:
        c[x] = c.get(x, 0) + 1

    max_c = max(c.values())

    a = []
    for x, count in c.items():
        if count == max_c:
            a.append(x)

    return min(a)

s = input()
l = list(map(int, s.split()))

rs = f(l)
print(rs)
```

```python
#W6A13
def s_to_d(s):
    l = s.split()
    d = {}

    for i in range(0, len(l), 2):
        k = l[i]
        v = l[i+1]
        d[k] = v

    return d

def invert(d):
    r = {}
    for k, v in d.items():
        r[v] = k
    return r

s = input()
d_in = s_to_d(s)
d_out = invert(d_in)
print(d_out)
```

```python
#W6A15
def s_to_d(s):
    l = s.split()
    d = {}

    for i in range(0, len(l), 2):
        k = l[i]
        v = l[i+1]
        d[k] = int(v)

    return d

def f(d, k):
    r = {}
    for key, val in d.items():
        if val > k:
            r[key] = val
    return r

s_in = input()
k = int(input())

d_in = s_to_d(s_in)
d_out = f(d_in, k)
print(d_out)
```
